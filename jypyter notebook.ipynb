# Feature Selection part

```python
movies.columns
```

```python
movies=movies[['id','title','overview','genres']]
```

```python
movies
```

```python
movies['tagline']=movies['overview']+movies['genres']
```

```python
movies
```

```python
new_data=movies.drop(columns=['overview','genres'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tagline'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="Spectre"].index[0]
```

```python
new_data.info()
```

```python
distance=sorted(list(enumerate(similarity[2])), reverse=True,key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movie):
    # Find the index of the movie in the dataset
    index = new_data[new_data['title'] == movie].index[0]
    
    # Calculate the similarity distances
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])
    
    # Print the top 5 recommended movies
    for i in distances[1:6]:  # Skipping the first one because it will be the movie itself
        print(new_data.iloc[i[0]].title)

# Call the function
recommend("Iron Man")

        
```

```python
recommend("Iron Man")
```
